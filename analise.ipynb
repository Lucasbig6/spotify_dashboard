```python
import pandas as pd
```

```python
df= pd.read_csv('spotify_data_clean.csv')
df
```

```python
# Criando um dicionário de tradução
traducao_colunas = {
    'track_id': 'ID da faixa',
    'track_name': 'Nome da faixa',
    'track_number': 'Número da faixa',
    'track_popularity': 'Popularidade da faixa',
    'explicit': 'Explícita',
    'artist_name': 'Nome do artista',
    'artist_popularity': 'Popularidade do artista',
    'artist_followers': 'Seguidores do artista',
    'artist_genres': 'Gêneros do artista',
    'album_id': 'ID do álbum',
    'album_name': 'Nome do álbum',
    'album_release_date': 'Data de lançamento',
    'album_total_tracks': 'Total de faixas no álbum',
    'album_type': 'Tipo de álbum',
    'track_duration_min': 'Duração da faixa (min)'
}


# Aplicando a tradução
df = df.rename(columns=traducao_colunas)
```
