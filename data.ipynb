```python
import kagglehub
from kagglehub import dataset_load, KaggleDatasetAdapter
import pandas as pd
import os
```

```python
# Baixar dataset
dataset_path = kagglehub.dataset_download("wardabilal/spotify-global-music-dataset-20092025")
```

```python
# Carregar usando kagglehub
df = dataset_load(
    KaggleDatasetAdapter.PANDAS,
    "wardabilal/spotify-global-music-dataset-20092025",
    "spotify_data clean.csv"
)
```

```python

# Salvar localmente
df.to_csv("spotify_data_clean.csv", index=False)
print("Arquivo salvo em spotify_data_clean.csv")
```
